```python
import boto3
def publish_to_sns(sub, msg):
    topic_arn = "<my sns arn>"
    sns = boto3.client("sns")
    response = sns.publish(
        TopicArn=topic_arn,
        Message=msg,
        Subject=sub
    )

def final_status(in_alertMessage,in_patient_name, in_admitted_ward, in_patient_age, in_patient_bp, in_patient_heartbeat):
    sub = "Patient Alerts"
    msg = """
        Alert Details -
        Alert Message: {alert_message}
        Patient Name: {patient_name}
        Admitted Ward: {admitted_ward}
        Age: {patient_age}
        BP: {patient_bp}
        Heart Beat: {patient_heartbeat}
        """.format(alert_message=in_alertMessage, patient_name=in_patient_name, admitted_ward=in_admitted_ward, patient_age=in_patient_age,patient_bp=in_patient_bp,patient_heartbeat=in_patient_heartbeat)
    publish_to_sns(sub, msg)
```
